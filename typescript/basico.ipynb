Trabalhando com funções em Typescript:

Formas de declarar funções e varaiveis recebidas por parametros

```typescript
function add (a:number, b = 5){
    return a + b
}


console.log(add(10))
console.log(add('10')) // erro de tipo 
console.log(add(10,6));
console.log(add(10,'6')); // erro de tipo



//OBS: Note como é DENO ele não irá representar o erro mas calculo estará errado.
```

Teste de função sem passar um valor constado no B

```typescript
function add (a: number, b: number){
    return a + b
}

console.log(add(10)) // erro falta variavel de parametro
console.log(add('10')) // erro de tipo 
console.log(add(10,6));
console.log(add(10,'6')); // erro de tipo
```
